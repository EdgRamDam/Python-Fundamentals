```
import pandas as pd
```

```
df = pd.read_csv("/dbfs/FileStore/shared_uploads/edgar.ramirezd@kof.com.mx/athlete_events.csv")
df = df[df["Medal"].notnull()]
df = pd.DataFrame(df.groupby(["NOC"]).size().reset_index(name="medal_count"))
df = df.sort_values(by="medal_count",ascending=False).head(10)
df = df.sort_values(by="medal_count",ascending = True)
df.display()
df.plot(kind="bar")
```
