```
edad = int(dbutils.widgets.get("edad"))
sexo = dbutils.widgets.get("sexo")
mayorEdad=""
genero=""

if edad >= 18:
    mayorEdad="mayor"
else:
    mayorEdad="menor"
#
if sexo == "Hombre":
    genero="masculino"
else:
    genero="femenino"
#

print("Eres {0} de edad, género {1}".format(mayorEdad, genero))
```
