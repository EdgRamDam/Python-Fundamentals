```
import pandas as pd
import numpy as np
```

```
df_estudiantes = pd.read_csv("/dbfs/FileStore/shared_uploads/edgar.ramirezd@kof.com.mx/clean_students_complete.csv")
df_estudiantes.set_index("Student ID",inplace=True)
df_estudiantes =  df_estudiantes.drop(["Unnamed: 0"],axis=1)
```

```
#Crea / Presenta la forma que creas más conveniente para describir cuántos alumnos por escuela tuvieron el mejor puntaje en "reading", y con ello saber que género es el que predomina en este filtro.
df_mejor_reading = df_estudiantes[df_estudiantes["reading_score"] == df_estudiantes["reading_score"].max()]
df_dummies_gender = pd.get_dummies(df_mejor_reading["gender"])
df_plot = pd.concat([df_mejor_reading["school_name"],df_dummies_gender],axis=1)
df_plot.groupby("school_name").sum().plot(kind="bar",figsize=(15,4))
```

```
#Crea / Presenta la forma que creas más conveniente para describir cuántos alumnos por escuela tuvieron el peor puntaje en "reading", y con ello saber que género es el que predomina en este filtro.
df_peor_reading = df_estudiantes[df_estudiantes["reading_score"] == df_estudiantes["reading_score"].min()]
df_dummies_gender = pd.get_dummies(df_peor_reading["gender"])
df_plot = pd.concat([df_peor_reading["school_name"],df_dummies_gender],axis=1)
df_plot.groupby("school_name").sum().plot(kind="bar",figsize=(15,4))
```
