```
import pandas as pd
```

```
df = pd.read_csv("/dbfs/FileStore/shared_uploads/edgar.ramirezd@kof.com.mx/athlete_events.csv")
df = df[df["Year"] == df["Year"].max()]
df = df[df["Medal"].notnull()]
df = df[["Sex","Medal"]]
df2 = pd.get_dummies(df[["Medal"]])
resultado = pd.concat([df,df2],axis=1)
s = resultado.groupby(["Sex"]).sum()
s["Total"] = s["Medal_Bronze"] + s["Medal_Gold"] + s["Medal_Silver"]
plot = s["Total"].plot.pie(figsize = (11,6))
fig = plot.get_figure()
fig.savefig("/dbfs/FileStore/shared_uploads/edgar.ramirezd@kof.com.mx/sesion16_pay.png")
```
