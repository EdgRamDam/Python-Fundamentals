```
import random
dicc = dict.fromkeys(["Felipe","Olivia","Gabriel","Cesar","Adolfo","Monserrat"])

for k in dicc.keys():
    dicc[k] = random.randint(1,100)
#

for k,v in dicc.items():
    print("A {0} se le asignó el número: {1}".format(k,v))
#

print("El número más grande es: {0}".format(max(dicc.values())))
print("El número más pequeño es: {0}".format(min(dicc.values())))
```
