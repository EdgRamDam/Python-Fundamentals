## 1. Importar librerías

```
import pandas as pd
import numpy as np
```

## 2. Cargar archivo en Data Frame, eliminar columna duplicada y setear ID de estudiante como índice

```
df_estudiantes = pd.read_csv("/dbfs/FileStore/shared_uploads/edgar.ramirezd@kof.com.mx/clean_students_complete.csv")
df_estudiantes.set_index("Student ID",inplace=True)
df_estudiantes =  df_estudiantes.drop(["Unnamed: 0"],axis=1)
```

## 3. Generar columnas categóricas para reading_score y math_score

```
df_estudiantes["cat_reading"] = pd.Categorical(df_estudiantes["reading_score"])
df_estudiantes["cat_math"] = pd.Categorical(df_estudiantes["math_score"])
df_estudiantes
```

## 4. Aplicar value_counts a columnas categóricas y generar gráfico

```
df_estudiantes[["cat_reading","cat_math"]].apply(pd.Series.value_counts).plot.bar(figsize=(25,10))
```
