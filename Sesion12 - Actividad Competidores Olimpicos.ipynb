```
import pandas as pd
```

```
data = pd.read_csv("/dbfs/FileStore/shared_uploads/edgar.ramirezd@kof.com.mx/athlete_events.csv")
```

```
data.display()
```

```
#Crear un programa en Visual Studio que me permita saber cul es el competidor ms veterano que ha recibido medalla para los NOCs MEX, COL y ARG (oro, plata o bronce) 
data1 = data[data["NOC"].isin(["MEX","COL","ARG"]) & data["Medal"].notnull()]
data1[data1["Age"] == data1["Age"].max()].display()
```

```
#Crear un programa en Visual Studio que me permita saber cul es el competidor ms joven que ha recibido medalla de oro para los NOCs USA y CAN
data2 = data[data["NOC"].isin(["USA","CAN"]) & data["Medal"].isin(["Gold"])]
data2[data2["Age"] == data2["Age"].min()].display()
```

```
#Encuentra al competidor ms ganador de la historia en medallas totales, medallas de oro, medallas de plata y medallas de broce para los NOCs USA, CHN y RUS. Crea un csv con toda su informacin.
data3 = data[data["NOC"].isin(["USA","CHN","RUS"]) & data["Medal"].notnull()]
data3Agg = pd.DataFrame(data3.groupby(["Name"]).size().reset_index(name = "medal_count"))
data3Agg[data3Agg["medal_count"] == data3Agg["medal_count"].max()].display()
```
