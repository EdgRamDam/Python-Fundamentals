```
import pandas as pd
```

```
dict_1={
    "Nombre":["Ringo","John","Paul","George","Yoko"],
    "Edad":[45,34,42,38,47],
    "Salario":[12000,14000,13000,11000,10000],
    "Genero":["M","M","M","M","F"]
}
```

```
df = pd.DataFrame(dict_1)
```

```
print ("La diferencia entre el salario más alto ({0}) y el más bajo ({1}) es: {2}".format(df["Salario"].max(), df["Salario"].min(), df["Salario"].max()-df["Salario"].min()))
```

```
print("El promedio de salarios es: {0}".format(df["Salario"].mean()))
print("La mediana de salarios es: {0}".format(df["Salario"].median()))
print("La desviación estándar de salarios es: {0}".format(df["Salario"].std()))
```
