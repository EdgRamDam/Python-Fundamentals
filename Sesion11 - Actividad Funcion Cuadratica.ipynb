```
import math
```

```
def fCuadratica (a,b,c):
    x1 = (-b + math.sqrt(b*b + 4*a*c)) / 2*a
    x2 = (-b - math.sqrt(b*b + 4*a*c)) / 2*a
    return x1, x2
#
```

```
a,b,c = 4,-5,6
resultado = fCuadratica(a,b,c)
print("Resultados de fCuadratica con a={0}, b={1}, c={2}:\nx1={3}\nx2={4}".format(a,b,c,resultado[0],resultado[1]))
```
